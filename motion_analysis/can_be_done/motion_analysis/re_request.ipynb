```python
import requests,json
import json
import os
g = os.walk("./raw_data/")

scores = []
reviews = []
review_scores = []
```

```python
def quest(PARAMS):
    URL = "https://sent-an.herokuapp.com/predict?q="
    url = URL + PARAMS
    r = requests.get(url)
    j = json.loads(str(r.text))
#     print(j['predicted_class'])
    return j['prediction'], j['predicted_class']

def walk_file(file_name, score, review, review_score):
    file_json = []
    with open(file_name, 'rb') as f:
#             line = f.readline()
            lines = [l.decode('utf8', 'ignore') for l in f.readlines()]
            json_data = json.loads(line)
            for d in json_data:
                data = {}
                data['score'] = d['score']
                data['review'] = d['review']
                predict, clazz = quest(d['review'])
                data['class'] = clazz
                data['prediction'] = predict
                
#                 score.append(d['score'])
#                 review.append(d['review'])
#                 review_score.append(quest(d['review']))
                file_json.append(data)
            f.close()
            file_name = os.path.splitext(file_name)[0]
            print(file_name+str(" done! \n"))
            filename = file_name+str('_res.json')
            save2json(file_json, filename)
            return score, review, review_score

def save2json(dictObj, filename):
    jsonObj = json.dumps(dictObj)
    fileObject = open(filename,'w')
    fileObject.write(jsonObj)
    fileObject.close()  
```

```python
for path, d, filelist in g:
    print(d)
    for filename in filelist:
        file_name = os.path.join(path, filename)
        score = []
        review = []
        review_score = []
        score, review, review_score = walk_file(file_name, score, review, review_score)
        scores.append(score)
        reviews.append(review)
        review_scores.append(review_score)
```
